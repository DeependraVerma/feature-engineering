```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
df["number"].unique()
```

```python
df.groupby("number").size().sort_values(ascending = False).plot(kind = "bar")
plt.show()
```

```python
# extract numerical part
df["number_numerical"] = pd.to_numeric(df["number"],errors="coerce",downcast="integer")
```

```python
# extract categorical part
df["number_categorical"] = np.where(df["number_numerical"].isnull(),df["number"],np.nan)
```

```python
df
```

```python
df["Cabin"].unique()
```

```python
# captures numerical part

df["cabin_num"] = df["Cabin"].str.extract("(\d+)")
df["cabin_cat"] = df["Cabin"].str[0]
```

```python
df
```

```python
df["Ticket"].unique()
```

```python
# extract the last bit of ticket as number
df["ticket_num"] = df["Ticket"].apply(lambda s:s.split()[-1])
df["ticket_num"] = pd.to_numeric(df["ticket_num"],errors = "coerce", downcast = "integer")
```

```python
df.info()
```

```python
# extract the first bit of ticket as category
df["ticket_category"] = df["Ticket"].apply(lambda s:s.split()[0])
df["ticket_category"]  = np.where(df["ticket_category"].str.isdigit(),np.nan,df["ticket_category"])
```
