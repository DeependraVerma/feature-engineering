```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("covid_toy.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop("has_covid", axis = 1),df["has_covid"], test_size=0.2, random_state= 0)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[("tnf1",SimpleImputer(),["fever"]),
                                              ("tnf2",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
                                              ("tnf3",OneHotEncoder(sparse = False,drop = "first"),["gender","city"])], remainder="passthrough")
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.fit_transform(X_test)
```
